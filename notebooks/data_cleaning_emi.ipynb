## Correcting data types

```python
df['age'] = df['age'].str.replace('.0.0', '.0')
df['age'] = df['age'].astype('float')
df['age'].value_counts()
```

```python
df['monthly_salary'] = df['monthly_salary'].str.replace('.0.0', '.0')
df['monthly_salary'] = df['monthly_salary'].astype('float')
df['monthly_salary'].value_counts()

```

```python
df['bank_balance'] = df['bank_balance'].str.replace('.0.0', '.0').str.replace('nan.0', 'nan')
df['bank_balance'] = df['bank_balance'].astype('float')
df['bank_balance'].value_counts()
```
